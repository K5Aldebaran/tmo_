## Цель работы
Провести корреляционный анализ набора данных о химическом составе вин (Wine dataset, UCI). Обработать искусственно созданные пропуски, оценить взаимосвязи признаков и сделать выводы о возможности построения моделей машинного обучения.

```python
import pandas as pd
import numpy as np

# Ссылка на файл с данными о вине
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"

# Названия столбцов (что означает каждый столбец)
columns = ['Class', 'Alcohol', 'Malic_acid', 'Ash', 'Alcalinity_of_ash', 'Magnesium',
           'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols', 'Proanthocyanins',
           'Color_intensity', 'Hue', 'OD280_OD315', 'Proline']

# Загружаем данные в таблицу
df = pd.read_csv(url, header=None, names=columns)

# Показываем первые 5 строк таблицы
print("Первые 5 строк данных:")
print(df.head())

# Показываем размер таблицы (строки, столбцы)
print(f"\nРазмер таблицы: {df.shape[0]} строк, {df.shape[1]} столбцов")
```

```python
# Считаем пропуски в каждом столбце
print("Сколько пропусков в каждом столбце:")
print(df.isnull().sum())
```

```python
import random

# Фиксируем random seed, чтобы результат был одинаковым при каждом запуске
random.seed(42)
np.random.seed(42)

# Копируем таблицу, чтобы не испортить оригинал
df_with_nan = df.copy()

# В столбце "Magnesium" заменяем 5% строк на пропуски
magnesium_mask = np.random.random(len(df_with_nan)) < 0.05
df_with_nan.loc[magnesium_mask, 'Magnesium'] = np.nan

# В столбце "Flavanoids" заменяем 5% строк на пропуски
flavanoids_mask = np.random.random(len(df_with_nan)) < 0.05
df_with_nan.loc[flavanoids_mask, 'Flavanoids'] = np.nan

# Проверяем, сколько теперь пропусков
print("Пропуски после искусственного создания:")
print(df_with_nan.isnull().sum())
```

```python
# Удаляем все строки, где есть хотя бы один пропуск
df_clean = df_with_nan.dropna()

print(f"Было строк: {len(df_with_nan)}")
print(f"Стало строк: {len(df_clean)}")
print(f"Удалено строк: {len(df_with_nan) - len(df_clean)}")
```

```python
# Считаем корреляции между всеми столбцами
correlation_matrix = df_clean.corr()

# Показываем корреляцию с сортом вина (Class)
print("Корреляция каждого признака с сортом вина (Class):")
corr_with_class = correlation_matrix['Class'].drop('Class').sort_values(ascending=False)
print(corr_with_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Настраиваем размер картинки
plt.figure(figsize=(12, 10))

# Рисуем тепловую карту
sns.heatmap(correlation_matrix,
            annot=True,      # показываем числа внутри клеток
            fmt='.2f',       # 2 знака после запятой
            cmap='coolwarm', # цвет: красный(полож) - синий(отриц)
            square=True,     # квадратные клетки
            linewidths=0.5)  # тонкие линии между клетками

# Добавляем заголовок
plt.title('Матрица корреляций признаков вина', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Берём 3 самых сильных признака (по модулю)
top_3 = corr_with_class.abs().nlargest(3)

print("ТОП-3 признака, влияющих на сорт вина:")
print(top_3)

# Строим графики для этих признаков
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, (feature, corr) in enumerate(top_3.items()):
    ax = axes[i]
    # Для каждого сорта вина (1,2,3) рисуем точки
    for wine_class in [1, 2, 3]:
        subset = df_clean[df_clean['Class'] == wine_class]
        ax.scatter(subset[feature], subset['Class'],
                   label=f'Сорт {wine_class}', alpha=0.7)
    ax.set_xlabel(feature)
    ax.set_ylabel('Сорт вина')
    ax.set_title(f'{feature}\n(корреляция = {corr:.3f})')
    ax.set_yticks([1, 2, 3])
    ax.legend()

plt.suptitle('Как признаки влияют на сорт вина', fontsize=14)
plt.tight_layout()
plt.show()
```

### 3. Можно ли строить модели машинного обучения?

**да, можно**

Почему:
- Признаки хорошо коррелируют с целевой переменной (сортом)
- Есть чёткие закономерности
- Данные чистые (после удаления пропусков)

**Какие модели лучше использовать:**
- Random Forest (не боится сложных связей)
- Логистическая регрессия (нужна регуляризация)
- KNN (хорошо работает при малом количестве данных)

### 4. Что мешает
- Есть мультиколлинеарность (некоторые признаки сильно связаны между собой, например Flavanoids и Total_phenols)
- Мало данных (всего 178 строк)

### 5. Итог
Задача классификации вина по химическому составу решается легко. Даже простая модель даст точность >95%.

```python
# 1. Проверка на нормальность (для обоснования корреляции Пирсона)
from scipy.stats import shapiro
print("\n1. Проверка на нормальность распределения (тест Шапиро-Уилка):")
for col in ['Flavanoids', 'OD280_OD315', 'Proline']:
    stat, p = shapiro(df_clean[col].dropna())
    print(f"   {col}: p-value = {p:.4f} {'(нормально)' if p > 0.05 else '(не нормально)'}")

# 2. Корреляция Спирмена (для робастности)
print("\n2. Корреляция Спирмена (ранговая) с Class:")
spearman_corr = df_clean.corr(method='spearman')['Class'].drop('Class').abs().nlargest(3)
print(spearman_corr)

# 3. Вывод:
print("\n3. Итоговое заключение по корреляции:")
print("   Основные выводы совпадают для Пирсона и Спирмена -> результаты надёжны.")
```
